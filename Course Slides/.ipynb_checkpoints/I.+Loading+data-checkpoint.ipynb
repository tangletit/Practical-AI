```python
%config IPCompleter.greedy=True
import pandas as pd
import numpy as np
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
olympia["medals"].describe()
```

Let's look at Switzerland

```python
che = olympia[olympia["iso3"] == "CHE"]
che
```

```python
che.set_index("year", inplace=True)
che
```

```python
che.loc[2016]
```

```python
che["medals"].plot(kind='bar')
```

```python
print("Switzerland got a median of %i medals" % che["medals"].median())
```

Does this mean Switzerland is above/below average (median)? --> need median for each country

```python
countries = olympia.groupby('country')
countries.get_group('Iran')
medal_median_per_country = countries[["medals","host"]].median()
medal_median_per_country.head(10)
```

Let's look at how these numbers are distributed

```python
medal_median_per_country["medals"].hist(bins=20)
```

most countries have between 0 and 5 medals.. So, does CHE score more medals than average? 

of all the countries that did win a medal. How did CHE fare there? 

```python
medal_median_per_country[medal_median_per_country["medals"] > 0]["medals"].median()
```

Looks like CHE wasn't that bad.
